```python
import pyautogui
import pandas as pd
from time import sleep

# Pandas pra ler a tabela do excle e puxar a coluna A da planilha
planilha = pd.read_excel('Relatório\Relação de Empresas por Franquia.xlsx')
Base_codigo = planilha.iloc[:, 0].dropna().astype(int).tolist()

ativo =  ['Ativowhite.png', 'Ativoblack.png']
inativo = ['Inativowhite.png', 'Inativoblack.png']
editar = ['Editar.png', 'Editar2.png']

def puxar(lista):
    for img in lista:
        pos = pyautogui.locateOnScreen(img, confidence=0.8)
        if pos:
            return pos
    return None

# verifica imagens
empresa_ativa = puxar(ativo)
empresa_inativa = puxar(inativo)
editar_empresa = puxar(editar)

# ir para home
pyautogui.click('Home1.png')
sleep(2)

# abrir busca
pyautogui.press('f8')
sleep(4)

codigo = Base_codigo[0]

# escrever código atual
pyautogui.write(str(codigo))
sleep(2)

pyautogui.click('Dados da empresa2.png')
sleep(5)

if empresa_ativa:
    print(f"{codigo} -> Empresa Ativa!")
    pyautogui.click('Ativoblack.png')
    sleep(2)
    pyautogui.click('Inativar Empresa.png')
    sleep(2)
    pyautogui.click('Gravar.png')
    sleep(5)
elif empresa_inativa:
    print(f"{codigo} -> Empresa Inativa!")
else:
    print(f"{codigo} -> Outro situação informada.")
```
