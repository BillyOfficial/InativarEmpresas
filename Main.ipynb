```python
import pyautogui
from time import sleep


Base_codigo = [2506, 2615, 2616]
ativo = ['Empresa ativa2.png', 'Empresa ativa.png']

pyautogui.click('Home1.png')
sleep(2)

pyautogui.press('f8')
sleep(4)

pyautogui.write(str(Base_codigo[0]))
sleep(2)

pyautogui.click('Dados da empresa2.png')
sleep(5)

for codigo in Base_codigo:
    
    achou = pyautogui.locateOnScreen(ativo, confidence=0.8)

    if achou:
        print("Empresa já inativa.")
        pyautogui.screenshot(f'{codigo}.png')
    else:
        pyautogui.click(ativo)
        sleep(2)
        pyautogui.click('Inativar Empresa.png')
        sleep(2)
        pyautogui.click('Gravar.png')
        sleep(7)
```
